## Filling Null Values Using fillna()

```python
data['total_sqft'] = data['total_sqft'].fillna(0)
print("DataFrame with Missing Values Filled:\n", data)
```

#### Filling Null Values Using fillna(method = 'pad')

```python
#Filling Null Values Using fillna(method = 'pad')
df_filled = data.fillna(method='pad')
print("Missing Values Filled (Forward Fill):\n", df_filled)
```

#### ● Filling Null Values Using fillna(method = 'bfill')

```python
#● Filling Null Values Using fillna(method = 'bfill')
df_filled = data.fillna(method='bfill')
print("Missing Values Filled (bacward Fill):\n", df_filled)
```

### Filling Null Values with the Mean, Max or Min of a Column

```python
#Mean
mean_value = data['price'].mean()
data['price'] = data['price'].fillna(mean_value)
mean_value
```

```python
mean_bath = data['bath'].mean()
# Fill missing values in the 'bath' column with the mean
data['bath'].fillna(mean_bath, inplace=True)

#mean_bath = data['bath'].mean()
#data['bath'] = data['bath'].fillna(mean_bath)

# Display the missing values filled
print(data)
```

```python
#Filling the maximum value
max_balcony = data['balcony'].max()

# Fill missing values in the 'balcony' column with the maximum value
data['balcony'].fillna(max_balcony, inplace=True)

# Display the missing values filled
print(data)
```

```python
#Fill the minimum value
min_price = data['price'].min()

# Fill missing values in the 'price' column with the minimum value
data['price'].fillna(min_price, inplace=True)

# Display the missing values filled
print(data)
```

### Dropping Null Values Using dropna() , dropna(how)

```python
# Drop columns with any null values
df_cleaned_columns = data.dropna(axis=1)

# Display the columns containing null values dropped
print(df_cleaned_columns)
```

```python
#dropna(how)
# Drop rows containing all null values
df_cleaned_all = data.dropna(how='all')

# Display the rows containing all null values dropped
print(df_cleaned_all)
```

### Filling Null Values Using replace()

```python
# Replace null values in a specific column (e.g., 'bath') with a specific value (e.g., -1)
data['bath'] = data['bath'].replace(np.nan, -1)

# Display the null values replaced in the specified column
print(data)
```

### Filling Null Values Using interpolate()

```python
# Fill null values using linear interpolation 
data_interpolated = data.interpolate()

# Display the DataFrame with null values interpolated
print(data_interpolated)
```
